```python
import pandas as pd
import matplotlib.pyplot as plt

# 1. Chargement des données
df = pd.read_csv(r"C:\Users\arbi\Desktop\PROJET2\covid_19_clean_complete.csv")

# 2. Nettoyage des données

# Supprimer doublons
df = df.drop_duplicates()

# Voir les colonnes et leurs types
print(df.info())

# Conversion de la colonne Date en datetime
df['Date'] = pd.to_datetime(df['Date'])

# Gestion des valeurs manquantes
print(df.isnull().sum())

# Pour ce dataset, il est probable que certaines colonnes comme 'Province/State' soient souvent vides, on peut les laisser.  
# Supprimer les lignes avec des valeurs manquantes dans les colonnes critiques (Confirmed, Deaths, Recovered)
df = df.dropna(subset=['Confirmed', 'Deaths', 'Recovered'])

# 3. Agrégation des données par pays et date (certains pays ont plusieurs provinces)
df_agg = df.groupby(['Country/Region', 'Date']).agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum'
}).reset_index()

# 4. Analyse rapide

# Exemple : afficher les 10 pays avec le plus de cas confirmés au dernier jour du dataset
last_date = df_agg['Date'].max()
df_last = df_agg[df_agg['Date'] == last_date]
top_countries = df_last.sort_values(by='Confirmed', ascending=False).head(10)
print("Top 10 pays au dernier jour enregistré :")
print(top_countries[['Country/Region', 'Confirmed']])

# Visualisation simple des cas confirmés pour ces 10 pays
plt.figure(figsize=(12,6))
plt.bar(top_countries['Country/Region'], top_countries['Confirmed'])
plt.title('Top 10 pays avec le plus de cas confirmés au {}'.format(last_date.date()))
plt.ylabel('Nombre de cas confirmés')
plt.xticks(rotation=45)
plt.show()

# 5. Préparation pour modélisation

# Exemple : choisir un pays (France ici) pour prédiction
country = 'France'
df_country = df_agg[df_agg['Country/Region'] == country].copy()

# On s'assure que les dates soient bien ordonnées
df_country = df_country.sort_values('Date')

# Créer des features temporelles qui peuvent aider pour la prédiction
df_country['Day'] = (df_country['Date'] - df_country['Date'].min()).dt.days

# Afficher un aperçu
print(df_country.head())

# 6. Export des données nettoyées pour la prédiction
df_country.to_csv(r"C:\Users\arbi\Desktop\PROJET2\covid_{}_cleaned.csv".format(country.lower()), index=False)
```

```python
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt
import numpy as np

# 1. Charger les données nettoyées pour la France (issue de notre traitement précédent)
df = pd.read_csv(r"C:\Users\arbi\Desktop\PROJET2\covid_france_cleaned.csv")

# 2. Préparer les données
X = df[['Day']]  # variable indépendante (jours depuis début)
y = df['Confirmed']  # variable cible (cas confirmés)

# 3. Diviser en train/test (ex: 80% train, 20% test)
split = int(0.8 * len(df))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# 4. Créer et entraîner le modèle de régression linéaire
model = LinearRegression()
model.fit(X_train, y_train)

# 5. Faire des prédictions
y_pred = model.predict(X_test)

# 6. Évaluer le modèle
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f"Mean Squared Error: {mse:.2f}")
print(f"R^2 Score: {r2:.2f}")

# 7. Visualiser les résultats
plt.figure(figsize=(10,6))
plt.scatter(X_test, y_test, color='blue', label='Données réelles')
plt.plot(X_test, y_pred, color='red', linewidth=2, label='Prédictions')
plt.xlabel('Jour depuis début')
plt.ylabel('Cas confirmés')
plt.title('Prédiction des cas confirmés COVID-19 en France (Régression Linéaire)')
plt.legend()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.arima.model import ARIMA
from sklearn.metrics import mean_squared_error

# Charger et préparer les données
df = pd.read_csv(r"C:\Users\arbi\Desktop\PROJET2\covid_19_clean_complete.csv")
df['Date'] = pd.to_datetime(df['Date'])

# Agréger par pays et date
df_agg = df.groupby(['Country/Region', 'Date']).agg({'Confirmed': 'sum'}).reset_index()

# Filtrer pour la France
country = 'France'
df_country = df_agg[df_agg['Country/Region'] == country].copy()
df_country = df_country.set_index('Date').sort_index()

# Séparer les données en train/test
train_size = int(len(df_country)*0.8)
train, test = df_country.iloc[:train_size], df_country.iloc[train_size:]

# Définir et ajuster le modèle ARIMA (p,d,q)
# p=5, d=1, q=0 sont des valeurs à tester/ajuster selon les données
model = ARIMA(train['Confirmed'], order=(5,1,0))
model_fit = model.fit()

# Faire des prédictions sur test
start = test.index[0]
end = test.index[-1]
predictions = model_fit.predict(start=start, end=end, typ='levels')

# Évaluer
mse = mean_squared_error(test['Confirmed'], predictions)
print(f'Mean Squared Error: {mse:.2f}')

# Visualiser
plt.figure(figsize=(12,6))
plt.plot(train.index, train['Confirmed'], label='Train')
plt.plot(test.index, test['Confirmed'], label='Test')
plt.plot(predictions.index, predictions, label='Prédictions ARIMA', color='red')
plt.title('Prédiction des cas confirmés COVID-19 en France avec ARIMA')
plt.xlabel('Date')
plt.ylabel('Cas confirmés')
plt.legend()
plt.show()
```

```python
import pandas as pd
import numpy as np
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_squared_error, r2_score
import matplotlib.pyplot as plt

# 1. Charger les données
df = pd.read_csv(r"C:\Users\arbi\Desktop\PROJET2\covid_19_clean_complete.csv")

# 2. Nettoyage de base
df = df.drop_duplicates()
df['Date'] = pd.to_datetime(df['Date'])

# 3. Agréger par pays et date
df_agg = df.groupby(['Country/Region', 'Date']).agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum'
}).reset_index()

# 4. Filtrer le pays (ex: France)
country = 'France'
df_country = df_agg[df_agg['Country/Region'] == country].copy()
df_country = df_country.sort_values('Date')

# 5. Créer une variable "Day" (jours depuis début)
df_country['Day'] = (df_country['Date'] - df_country['Date'].min()).dt.days

# 6. Préparer features et cibles multi-sortie
X = df_country[['Day']]
y = df_country[['Confirmed', 'Deaths', 'Recovered']]

# 7. Diviser en train/test (80/20)
split = int(0.8 * len(df_country))
X_train, X_test = X[:split], X[split:]
y_train, y_test = y[:split], y[split:]

# 8. Créer et entraîner le modèle Random Forest multi-sortie
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, y_train)

# 9. Prédire sur test
y_pred = model.predict(X_test)

# 10. Évaluer la performance pour chaque cible
for i, target in enumerate(['Confirmed', 'Deaths', 'Recovered']):
    mse = mean_squared_error(y_test.iloc[:, i], y_pred[:, i])
    r2 = r2_score(y_test.iloc[:, i], y_pred[:, i])
    print(f"{target} - MSE: {mse:.2f}, R^2: {r2:.2f}")

# 11. Visualiser les prédictions vs réels pour les cas confirmés
plt.figure(figsize=(12,8))
plt.subplot(3,1,1)
plt.plot(X_test, y_test['Confirmed'], label='Réel Confirmed')
plt.plot(X_test, y_pred[:,0], label='Prédit Confirmed', linestyle='--')
plt.legend()
plt.title('Cas confirmés')

plt.subplot(3,1,2)
plt.plot(X_test, y_test['Deaths'], label='Réel Deaths')
plt.plot(X_test, y_pred[:,1], label='Prédit Deaths', linestyle='--')
plt.legend()
plt.title('Décès')

plt.subplot(3,1,3)
plt.plot(X_test, y_test['Recovered'], label='Réel Recovered')
plt.plot(X_test, y_pred[:,2], label='Prédit Recovered', linestyle='--')
plt.legend()
plt.title('Récupérés')

plt.xlabel('Jour depuis début')
plt.tight_layout()
plt.show()

# 12. Prédictions pour les 7 prochains jours
future_days = 7
last_day = df_country['Day'].max()
future_X = np.array([last_day + i for i in range(1, future_days + 1)]).reshape(-1, 1)

future_pred = model.predict(future_X)

print("\nPrédictions pour les 7 prochains jours :")
for i in range(future_days):
    print(f"Jour +{i+1}: Confirmed={int(future_pred[i,0])}, Deaths={int(future_pred[i,1])}, Recovered={int(future_pred[i,2])}")
```
